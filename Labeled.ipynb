```python
import pandas as pd
```

```python
df = pd.read_csv (r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\test_amazon_en.csv')
df
```

```python
df['Label'] = df['Label'].str.replace('UNK', 'GR')
```

```python
item_counts = df['Label'].value_counts()
item_counts
```

```python
df_gr = df.loc[df['Label'] == 'GR']
df_gr
```

```python
df_gr = df_gr.sample(n=100)
df_gr
```

```python
df_sr = df.loc[df['Label'] == 'SR']
df_sr
```

```python
d = [df_gr, df_sr]
df = pd.concat(d)
df
```

```python
df = df.sample(n=200)
df
```

```python
test_data = df.sample(frac=0.8, random_state=25)
labeled_data = df.drop(test_data.index)

print(f"No. of test examples: {test_data.shape[0]}")
print(f"No. of labeled examples: {labeled_data.shape[0]}")
```

```python
df_test = test_data
df_labeled = labeled_data
```

```python
# Exporting the test dataset as a CSV file
df_test.to_csv(r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\Final Datasets\test_amazon.csv', index=False)
```

```python
# Exporting the labeled dataset as a CSV file
df_labeled.to_csv(r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\Final Datasets\labeled_amazon.csv', index=False)
```
