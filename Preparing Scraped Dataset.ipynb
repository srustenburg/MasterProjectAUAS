### Preparing scraped datset from Amazon

In this notebook, the dataset of English reviews which was created in the previous 'cleaning scraped dataset' notebook will be prepared to be used with GAN-BERT.

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading dataset
df = pd.read_csv (r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\amazon_en.csv')
df
```

### Dividing the dataset

```python
# Decreasing the amount of rows so the created test dataset is not too large to be manually labeled for GAN-BERT
df = df.sample(n=1000)
```

```python
# Creating a unlabeled and test dataset out of the entire dataset
unlabeled_data = df.sample(frac=0.8, random_state=25)
test_data = df.drop(unlabeled_data.index)

print(f"No. of unlabled examples: {unlabeled_data.shape[0]}")
print(f"No. of test examples: {test_data.shape[0]}")
```

```python
df_unlabeled = unlabeled_data
df_test = test_data
```

### Exporting the datasets as CSV files

```python
# Exporting the unlabeled dataset as a CSV file
df_unlabeled.to_csv(r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\Final Datasets\unlabeled_amazon.csv', index=False)
```

```python
# Exporting the test dataset as a CSV file for manual labeling
df_test.to_csv(r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\Final Datasets\test_amazon.csv', index=False)
```

### Reading the test dataset after manual labeling

After the initial test dataset was exported as a CSV file it was manually labeled using Microsft Excel. The now labeled test dataset, which is still located in the same directory, will now be reread in order to be further divided into a final test and labeled dataset.

```python
# Reading dataset
df = pd.read_csv (r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\test_amazon_en.csv')
df
```

```python
# Checking the distribution of values in the 'Label' column
item_counts = df['Label'].value_counts()
item_counts
```

```python
# Creating a test and labled dataset out of the initial test dataset
test_data = df.sample(frac=0.8, random_state=25)
labeled_data = df.drop(test_data.index)

print(f"No. of test examples: {test_data.shape[0]}")
print(f"No. of labeled examples: {labeled_data.shape[0]}")
```

```python
df_test = test_data
df_labeled = labeled_data
```

### Exporting the final test and labeled datasets as CSV files

```python
# Exporting the test dataset as a CSV file
df_test.to_csv(r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\Final Datasets\test_amazon.csv', index=False)
```

```python
# Exporting the labeled dataset as a CSV file
df_labeled.to_csv(r'C:\Users\Gebruiker\Desktop\Master Project\Datasets\Eigen Datasets\Final Datasets\labeled_amazon.csv', index=False)
```
